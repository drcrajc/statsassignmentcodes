```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_value = 78
threshold_prob = 0.005

# Calculate probabilities
k_values = np.arange(0, 200)  # Initial range for k values
probabilities = poisson.pmf(k_values, lambda_value)

# Find the maximum k where probability is above the threshold
max_k = np.max(k_values[probabilities >= threshold_prob])

# Update k_values and probabilities to reflect only up to max_k
k_values = np.arange(0, max_k + 1)
probabilities = poisson.pmf(k_values, lambda_value)

# Calculate expectation and median
expectation = lambda_value
median = poisson.median(lambda_value)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(k_values, probabilities, alpha=0.6, color='b', label='Probability Mass Function')
plt.axvline(expectation, color='r', linestyle='dashed', linewidth=1, label=f'Expectation (mean) = {expectation}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label=f'Median = {median}')

# Adding labels and title
plt.xlabel('Number of Meteorites')
plt.ylabel('Probability')
plt.title('Poisson Distribution of Meteorite Falls into the Ocean')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.ylim(1e-10, 1)  # Adjust y-scale for better visibility of small probabilities

# Show plot
plt.show()

(expectation, median)
```
